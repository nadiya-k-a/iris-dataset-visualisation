**DATA VISUALISATION**

*Species count*

*Comparison between various species based on sepal lenght and width*

```python
plt.figure(figsize=(17,9))
iris.plot(kind ='scatter',x='SL',y='SW',alpha=0.4,s=100*iris['PL'],c= 'hotpink')
```

```python
sns.jointplot(x='SL',y='SW',data = iris,size=5)
```

```python
##plot tht classsification data using a box plot

sns.boxplot(x='Classification',y='PL',palette ='husl',data= iris)
```

```python
##shows difference in the sum of sepal length for each species


data= iris.groupby('Classification')['SL'].sum().to_frame().reset_index()
plt.bar(data['Classification'],data['SL'],color=['cyan','red','blue'])
plt.title('Bar chart')
plt.xlabel('Classificaton')
plt.ylabel('Sepal length')
plt.show()
```
